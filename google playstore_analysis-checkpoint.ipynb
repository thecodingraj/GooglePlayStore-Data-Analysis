```python
print(df['Last Updated'].dtype)  # Should print: datetime64[ns]
```

```python
df['Size'] = df.Size.replace("Varies with device",np.nan)
df['Size'] = df.Size.str.replace("M","000") # All size values became the kilobyte type.
df['Size'] = df.Size.str.replace("k","")
df['Size'] = df.Size.replace("1,000+",1000)
df['Size'] =df['Size'].astype(float)
df['Size'].dtype
```

```python
# conver the size column to float 

def convert_size(size):
    if isinstance(size, str):
         if 'M' in size:
             return float(size.replace('M',''))
         elif 'k' in size:
             return float(size.replace('k',''))/1024
         else:
             return np.nan
df['Size']=df['Size'].apply(convert_size)
```

```python
# conver the installs column to int to remove +

df['Installs'] = df.Installs.str.replace(",","")
df['Installs'] = df.Installs.str.replace("+","")
df['Installs'] = df.Installs.replace("Free",np.nan)
df['Installs'] = df['Installs'].astype(float)
df['Installs'].dtype
```

# Task 5: Time Series Line Chart

```python
df['Installs'] = df['Installs'].str.replace(',', '')
df['Installs'] = df['Installs'].str.replace('+', '')
df['Installs'] = df['Installs'].replace('Free', 0)  
```

```python
df['Installs']=df['Installs'].astype(int)
```

```python
filtered_data = df[
    (df['Content Rating'] == 'Teen') &
    (df['App'].str.startswith('E')) &
    (df['Installs'] > 10_000)]
```

```python
from datetime import datetime, time

df['Last Updated'] = pd.to_datetime(df['Last Updated'])
```

```python
df['Month'] = df['Last Updated'].dt.to_period('M')
```

```python
monthly_installs = df.groupby(['Month', 'Category'])['Installs'].sum().reset_index()
```

```python
monthly_installs['MoM Growth'] = monthly_installs.groupby('Category')['Installs'].pct_change() * 100
```

```python
current_time = datetime.now().time()
start_time = time(18, 0)  
end_time = time(21, 0)    

if start_time <= current_time <= end_time:
    plt.figure(figsize=(12, 6))
    
    for category in monthly_installs['Category'].unique():
        category_data = monthly_installs[monthly_installs['Category'] == category]
        plt.plot(category_data['Month'].astype(str),category_data['Installs'],label=category)
        
        growth_periods = category_data[category_data['MoM Growth'] > 20]
        for _, row in growth_periods.iterrows():
            plt.fill_betweenx(
                [0, row['Installs']],
                row['Month'].start_time.strftime('%Y-%m'),
                row['Month'].end_time.strftime('%Y-%m'),
                color='green',
                alpha=0.3 )
    
    plt.title('Trend of Total Installs Over Time by Category')
    plt.xlabel('Time (Months)')
    plt.ylabel('Total Installs')
    plt.legend(title='Category')
    plt.xticks(rotation=45)
    plt.grid(True)
    
    plt.show()
else:
    print("The time series chart is only available between 6 PM and 9 PM IST.")
```
